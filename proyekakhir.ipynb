# Visualization & Explanatory Analysis

VISUALISASI kualitas udara kota Dingling dengan membuat dataframe yang menampilkan Jam, CO, dan O3 sebagai faktor penentu kualitas udara

```python
dingling = dingling_df_cleaned.groupby('hour')[['CO', 'O3']].mean().reset_index()

# Menampilkan DataFrame baru
print(dingling)
```

```python
plt.bar(dingling['hour'], dingling['CO'], label='CO', width=0.8)  # Plot CO
plt.bar(dingling['hour'], dingling['O3'], bottom=dingling['CO'], label='O3', width=0.6)  # Plot O3 di atas CO
plt.xlabel('jam')
plt.ylabel('kadar polutan CO dan O3')
plt.title('KADAR POLUTAN DI dingling')
plt.xticks(dingling['hour'])  # Atur label sumbu X dengan nilai 'hour'
plt.legend()  # Menampilkan legenda
plt.tight_layout()
plt.show()
```

VISUALISASI Kualitas udara kota Dongsi dengan membuat dataframe yang menampilkan Jam, CO, dan O3 sebagai faktor penentu kualitas udara

```python
dongsi = dongsi_df_cleaned.groupby('hour')[['CO', 'O3']].mean().reset_index()

# Menampilkan DataFrame baru
print(dongsi)
```

```python
plt.bar(dongsi['hour'], dongsi['CO'], label='CO', width=0.5)  # Plot CO
plt.bar(dongsi['hour'], dongsi['O3'], bottom=dongsi['CO'], label='O3', width=0.3)  # Plot O3 di atas CO
plt.xlabel('jam')
plt.ylabel('kadar polutan CO dan O3')
plt.title('KADAR POLUTAN DI DONGSI')
plt.xticks(dongsi['hour'])  # Atur label sumbu X dengan nilai 'hour'
plt.legend()  # Menampilkan legenda
plt.tight_layout()
plt.show()
```
